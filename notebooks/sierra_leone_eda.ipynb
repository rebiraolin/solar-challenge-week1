```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for plots
sns.set_style("whitegrid")

# Load the Sierra Leone dataset - IMPORTANT: Using the correct file name you provided
file_path_sierra_leone = '../data/sierraleone-bumbuna.csv'

try:
    df_sierra_leone = pd.read_csv(file_path_sierra_leone)
except FileNotFoundError:
    print(f"Error: {file_path_sierra_leone} not found. Please ensure the 'sierraleone-bumbuna.csv' file is in your 'data' folder.")
    raise # Re-raise to stop execution if file is missing

# Convert 'Timestamp' to datetime and sort
df_sierra_leone['Timestamp'] = pd.to_datetime(df_sierra_leone['Timestamp'])
df_sierra_leone = df_sierra_leone.sort_values(by='Timestamp')

print("DataFrame head:")
print(df_sierra_leone.head())
print("\nDataFrame info:")
df_sierra_leone.info()

---

# --- Summary Statistics and Missing Values ---
print("\nSummary Statistics for Numeric Columns:")
print(df_sierra_leone.describe())

# Missing Values Report
print("\nMissing Values Count:")
missing_values = df_sierra_leone.isnull().sum()
print(missing_values[missing_values > 0])

print("\nColumns with >5% Nulls:")
total_rows = len(df_sierra_leone)
for column, count in missing_values.items():
    if count / total_rows > 0.05:
        print(f"- {column}: {count} ({count/total_rows:.2%})")

# Check if 'Comments' column is entirely null
if 'Comments' in df_sierra_leone.columns and df_sierra_leone['Comments'].isnull().all():
    print("\n'Comments' column is entirely null and will be considered for removal.")

---

# --- Outlier Detection & Basic Cleaning ---
print("\n--- Outlier Detection & Basic Cleaning ---")

# Columns to check for outliers using Z-scores
outlier_cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Calculate Z-scores and flag rows with |Z|>3
for col in outlier_cols:
    if col in df_sierra_leone.columns:
        # Handle cases where std dev might be zero, preventing division by zero
        if df_sierra_leone[col].std() == 0:
            print(f"Warning: Standard deviation for '{col}' is zero. No Z-score calculated.")
            df_sierra_leone[f'Zscore_{col}'] = np.nan # Assign NaN or 0
        else:
            df_sierra_leone[f'Zscore_{col}'] = np.abs((df_sierra_leone[col] - df_sierra_leone[col].mean()) / df_sierra_leone[col].std())
            outliers = df_sierra_leone[df_sierra_leone[f'Zscore_{col}'] > 3]
            print(f"\nPotential outliers in '{col}' (|Z|>3): {len(outliers)} rows")
            if not outliers.empty:
                print(outliers[[col, 'Timestamp', f'Zscore_{col}']].head())
    else:
        print(f"Warning: Column '{col}' not found for outlier detection.")


# Basic Cleaning: Dropping the 'Comments' column if it's entirely null
if 'Comments' in df_sierra_leone.columns and df_sierra_leone['Comments'].isnull().all():
    df_sierra_leone = df_sierra_leone.drop(columns=['Comments'])
    print("\n'Comments' column dropped as it contained only null values.")
else:
    print("\n'Comments' column not dropped (not entirely null or not found).")


# Handling other missing values: (If any were found in initial check, add imputation or dropping logic here)
print("\nMissing values in key columns (after initial check, should be zero or very few):")
print(df_sierra_leone[outlier_cols].isnull().sum())

# Export cleaned DataFrame
import os
output_dir = '../data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

output_path_sierra_leone_clean = os.path.join(output_dir, 'sierra_leone_clean.csv')
df_sierra_leone.to_csv(output_path_sierra_leone_clean, index=False)
print(f"\nCleaned Sierra Leone data exported to: {output_path_sierra_leone_clean}")
```
