```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load cleaned data for each country
df_benin_clean = pd.read_csv('../data/benin_clean.csv', parse_dates=['Timestamp'])
df_sierra_leone_clean = pd.read_csv('../data/sierra_leone_clean.csv', parse_dates=['Timestamp'])
df_togo_clean = pd.read_csv('../data/togo_clean.csv', parse_dates=['Timestamp'])

print("Data loaded successfully!")
print(f"Benin Cleaned Data Shape: {df_benin_clean.shape}")
print(f"Sierra Leone Cleaned Data Shape: {df_sierra_leone_clean.shape}")
print(f"Togo Cleaned Data Shape: {df_togo_clean.shape}")
```

```python
# Calculate daily average GHI for each country
df_benin_clean['Date'] = df_benin_clean['Timestamp'].dt.date
df_sierra_leone_clean['Date'] = df_sierra_leone_clean['Timestamp'].dt.date
df_togo_clean['Date'] = df_togo_clean['Timestamp'].dt.date

daily_avg_ghi_benin = df_benin_clean.groupby('Date')['GHI'].mean().reset_index()
daily_avg_ghi_sierra_leone = df_sierra_leone_clean.groupby('Date')['GHI'].mean().reset_index()
daily_avg_ghi_togo = df_togo_clean.groupby('Date')['GHI'].mean().reset_index()

print("\nAverage GHI per country (across all data points):")
print(f"Benin: {df_benin_clean['GHI'].mean():.2f}")
print(f"Sierra Leone: {df_sierra_leone_clean['GHI'].mean():.2f}")
print(f"Togo: {df_togo_clean['GHI'].mean():.2f}")

# Plotting daily average GHI (quick and simple)
plt.figure(figsize=(12, 6))
plt.plot(daily_avg_ghi_benin['Date'], daily_avg_ghi_benin['GHI'], label='Benin')
plt.plot(daily_avg_ghi_sierra_leone['Date'], daily_avg_ghi_sierra_leone['GHI'], label='Sierra Leone')
plt.plot(daily_avg_ghi_togo['Date'], daily_avg_ghi_togo['GHI'], label='Togo')
plt.title('Daily Average GHI Across Countries')
plt.xlabel('Date')
plt.ylabel('Average GHI')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
