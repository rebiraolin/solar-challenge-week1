```python
# Start of togo_eda.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load the Togo dataset
file_path_togo = '../data/togo-dapaong_qc.csv' # CORRECTED PATH AND FILENAME

try:
    df_togo = pd.read_csv(file_path_togo) # CHANGED DF NAME
except FileNotFoundError:
    print(f"Error: {file_path_togo} not found. Please check the file path.")
    raise

df_togo['Timestamp'] = pd.to_datetime(df_togo['Timestamp']) # CHANGED DF NAME
df_togo = df_togo.sort_values(by='Timestamp') # CHANGED DF NAME

print("DataFrame head:")
print(df_togo.head()) # CHANGED DF NAME
print("\nDataFrame info:")
df_togo.info() # CHANGED DF NAME

# ... (rest of the summary stats, missing values, outlier code) ...

# Export cleaned DataFrame (important for Task 3)
import os
output_dir = '../data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

output_path_togo_clean = os.path.join(output_dir, 'togo_clean.csv') # CHANGED PATH AND FILENAME
df_togo.to_csv(output_path_togo_clean, index=False) # CHANGED DF NAME
print(f"\nCleaned Togo data exported to: {output_path_togo_clean}") # CHANGED MESSAGE
```
